1. An exciting new halal restaurant just opened in Greenwich, but hasn't been rated yet. The magazine has asked you to include it in your analysis. Add the following restaurant "Penang Flavours" to the database.

```python
# Create a dictionary for the new restaurant data
newhalal = {
    "BusinessName":"Penang Flavours",
    "BusinessType":"Restaurant/Cafe/Canteen",
    "BusinessTypeID":"",
    "AddressLine1":"Penang Flavours",
    "AddressLine2":"146A Plumstead Rd",
    "AddressLine3":"London",
    "AddressLine4":"",
    "PostCode":"SE18 7DY",
    "Phone":"",
    "LocalAuthorityCode":"511",
    "LocalAuthorityName":"Greenwich",
    "LocalAuthorityWebSite":"http://www.royalgreenwich.gov.uk",
    "LocalAuthorityEmailAddress":"health@royalgreenwich.gov.uk",
    "scores":{
        "Hygiene":"",
        "Structural":"",
        "ConfidenceInManagement":""
    },
    "SchemeType":"FHRS",
    "geocode":{
        "longitude":"0.08384000",
        "latitude":"51.49014200"
    },
    "RightToReply":"",
    "Distance":4623.9723280747176,
    "NewRatingPending":True
}
```

```python
# Insert the new restaurant into the collection
establishments.insert_one(newhalal)
```

```python
# Check that the new restaurant was inserted
newrest = list(establishments.find(newhalal))
newrest
```

2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
#list(establishments.find_one({"BusinessType":"Restaurant/Cafe/Canteen"}))
```

```python
list(establishments.find({"BusinessType":"Restaurant/Cafe/Canteen"}))
```

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
rest = list(establishments.find({"BusinessType":"Restaurant/Cafe/Canteen"}))
restid = rest[0]["BusinessTypeID"]
restid
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID
establishments.update_many({"BusinessType":"Restaurant/Cafe/Canteen"},
                          {"$set":{"BusinessTypeID":restid}})
```

```python
# Confirm that the new restaurant was updated
list(establishments.find({'BusinessName': 'Penang Flavours'}))
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"
len(list(establishments.find({'LocalAuthorityName': 'Dover'})))
```

```python
establishments.count_documents({'LocalAuthorityName': 'Dover'})
```

```python
# Check that other documents remain with 'find_one'
establishments.find_one()
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
#.update_many({}, [])
#.update_many({}, [{"set": {}}])
```

```python
# Change the data type from String to Decimal for longitude and latitude
establishments.update_many({}, [{"$set": {"geocode.longitude": {"$toDouble": "$geocode.longitude"},
                                         "geocode.latitude": {"$toDouble": "$geocode.latitude"}}}])
```

```python
establishments.find_one()
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
establishments.distinct("RatingValue")
```

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ { '$set': {"RatingValue" : None} } ])
```

```python
establishments.distinct("RatingValue")
```

```python
# Change the data type from String to Integer for RatingValue
establishments.update_many({}, [{"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}])
```

```python
establishments.distinct("RatingValue")
```

```python
# Check that the coordinates and rating value are now numbers
establishments.find_one()
```
