```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query = {"$match": {"RatingValue": 5, 
                    "geocode.latitude": {"$gte":51.480142, 
                                         "$lte":51.500142},
                   "geocode.longitude": {"$gte":0.07384, 
                                         "$lte":0.09384}}}
sort = {"$sort": {"scores.Hygiene": -1}}
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest (number of documents?)

# Print the number of documents in the result

# Print the first 10 results
```

```python
query = {"$match": {"scores.Hygiene": 0}}
group = {"$group": {"_id": "$LocalAuthorityName",
                   "numberOfDocs": {"$sum":1}}}
sort = {"$sort": {"numberOfDocs": -1}}
```

```python
leasthyg = list(establishments.aggregate([query, group, sort]))[0:10]
leasthyg
```

```python
# Convert the result to a Pandas DataFrame
leasthyg_df = pd.DataFrame(leasthyg)

# Display the number of rows in the DataFrame
print("number of rows:", len(leasthyg_df))

# Display the first 10 rows of the DataFrame
leasthyg_df.columns = ["LocalAuthority", "Number"]
leasthyg_df
```
